```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

sys.path.append(os.path.abspath('../src'))
from noise import identity, add_gaussian_noise, add_uniform_noise, add_sinusoidal_noise, add_exponential_noise

q = 0.5
epsilon = 0.025

noisy_q_gaussian = [add_gaussian_noise(q, epsilon) for _ in range(10000)]
noisy_q_uniform = [add_uniform_noise(q, epsilon) for _ in range(10000)]
noisy_q_sinusoidal = [add_sinusoidal_noise(q, epsilon) for _ in range(10000)]
noisy_q_exponential = [add_exponential_noise(q, epsilon) for _ in range(10000)]

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(noisy_q_gaussian, bins=20, color='blue', alpha=0.7)
plt.title('Gaussian Noise')
plt.xlabel('Noisy q')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(noisy_q_uniform, bins=20, color='green', alpha=0.7)
plt.title('Uniform Noise')
plt.xlabel('Noisy q')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(noisy_q_sinusoidal, bins=20, color='red', alpha=0.7)
plt.title('Sinusoidal Noise')
plt.xlabel('Noisy q')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(noisy_q_exponential, bins=20, color='purple', alpha=0.7)
plt.title('Exponential Noise')
plt.xlabel('Noisy q')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
