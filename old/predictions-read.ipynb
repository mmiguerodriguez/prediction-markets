```python
import pandas as pd

def load_and_display_data(data_filename, predictions_filename):
  data_df = pd.read_csv(data_filename)
  predictions_df = pd.read_csv(predictions_filename)
  display(data_df)
  display(predictions_df)
```

```python
base = "results/"
data_filename = base + "scores.csv"
predictions_filename = base + "predictions.csv"
load_and_display_data(data_filename, predictions_filename)
```
